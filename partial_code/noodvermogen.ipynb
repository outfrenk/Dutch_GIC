```python
import numpy as np
import pandas as pd
path='/usr/people/out/Documents/noodvermogen'
olddate=None
indicator=False
TenneT=pd.read_csv(f'{path}/2003export.csv', delimiter=',')
TenneT.fillna(0)
Nood=pd.DataFrame(columns=['datum','starttijd','eindtijd','power'])

for line in range(len(TenneT)):
    if TenneT.at[line,'opregel noodvermogen'] > 0:
        date=TenneT.at[line,'datum']
        start=TenneT.at[line,'periode_van']
        if olddate!=date:
#             print(date)
            olddate=date
            oldstart=start
            power=TenneT.at[line,'opregel noodvermogen']
        elif olddate==date and indicator==False:
#             print(start)
            oldstart=start
            power=TenneT.at[line,'opregel noodvermogen']
        else:
            power+=TenneT.at[line,'opregel noodvermogen']
        oldline=line
        indicator=True
    if indicator and not TenneT.at[line,'opregel noodvermogen'] > 0:
        end=TenneT.at[oldline,'periode_tm']
#         print(end)
#         print()
        Nood=Nood.append(pd.Series([olddate, oldstart, end, power], index=Nood.columns),ignore_index=True)
        indicator=False
```
